```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Banking.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
bins = [0,100000,300000,float('inf')]
labels = ['low','Med','High']

df['Income_band'] = pd.cut(df['Estimated Income'],bins = bins , labels = labels, right = False)
```

```python
df['Income_band'].value_counts().plot(kind = 'bar')
```

```python
categorical_cols = df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Income_band','Risk Weighting']]

for col in categorical_cols:
    print(f"Value Counts for '{col}' : ")
    display(df[col].value_counts())
```

```python
#univariate analysis
for i,predicator in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(x = predicator,data = df)
```

```python
for i,predicator in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(x = predicator,data = df, hue = 'Nationality')
```

```python
for col in categorical_cols:
    if col=='Occupation':
        continue
    plt.figure(figsize=(10,5))
    sns.histplot(df[col],kde = True)
    plt.title("Histogram of occupation count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Checking Accounts','Bank Deposits','Foreign Currency Account','Business Lending']

plt.figure(figsize=(20,30))
for i ,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde= True)
    plt.title(col)
plt.show()
```

```python
corrr = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corrr, annot = True, cmap = 'crest', linewidths = 0.2)
plt.title("Correlation Heatmap")    
```
